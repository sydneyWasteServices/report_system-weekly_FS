```python
import pyodbc as odbc
import pandas as pd
import math
import numpy as np

df = pd.read_csv("../../dataVault/Dec_20/df_test1.csv") 

top_three_rows = df.head(300)

col_values = top_three_rows.columns.values
df = top_three_rows.replace(np.nan, None, regex=True)


print(df)

connStr = odbc.connect('DRIVER={ODBC Driver 11 for SQL Server};\
                        SERVER=SW-SRV1\ATSSQLEXPRESS;DATABASE=test_db;\
                        Trusted_Connection=yes')

cursor = connStr.cursor()

insert_query = ('INSERT INTO Finance.test_db_1('
        '[date],'
        '[Customer_Name],'
        '[Address],'
        '[City],'
        '[State],'
        '[PostCode],'
        '[Qty_Scheduled],'
        '[Qty_Serviced],'
        '[Serv_Type],'
        '[Bin_Volume],'
        '[Status],'
        '[Truck_number],'
        '[Route_number],'
        '[Waste_Type],'
        '[Price])'
        'values (CONVERT(DATETIME,?,103),?,?,?,?,?,?,?,?,?,?,?,?,?,?);')


row_count_query = "SELECT count(*) FROM Finance.test_db_1;"

def is_postCode_empty(postCode : int):
    if(postCode):
        print(postCode)
    else:
        return postCode

for index,row in df.iterrows():
    try:
            cursor.execute(insert_query, row['Date'],
            row['Customer Name'],
            row['Address 1'],
            row['City'],
            row['State'],
#             is_postCode_empty(row['PostCode']),
            row['PostCode'],
            row['Qty Scheduled'],
            row['Qty Serviced'],
            row['Serv Type'],
            row['Bin Volume'],
            row['Status'],
            row['Truck number'],
            row['Route number'],
            row['Waste Type'],
            row['Price'])
    
    except:
        print(row['PostCode'])
    connStr.commit()
        
total_rows  = pd.read_sql(row_count_query, connStr)

cursor.close()
connStr.close()

print(total_rows)
```
